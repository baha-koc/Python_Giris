```python
#BÖLÜM 4: OPERATÖRLER

sayi1 = 50
sayi2 = 15

#Aritmetik İşlemler
print(f"Toplam: {sayi1 + sayi2}")
print(f"Fark: {sayi1 - sayi2}")
print(f"Çarpım: {sayi1 * sayi2}")
print(f"Bölüm: {sayi1 / sayi2}")

#Karşılaştırma Operatörleri
girilen_sayi = int(input("Bir sayı girin: "))

print("100'den büyük mü?:", girilen_sayi > 100)
print("100'e eşit mi?:", girilen_sayi == 100)
print("100'den küçük mü?:", girilen_sayi < 100)

#Mantıksal Operatörler
kullanici_adi = input("Kullanıcı adı: ")
sifre = input("Şifre: ")

giris_basarili = (kullanici_adi == "admin") and (sifre == "12345")
print("Giriş sağlandı mı?:", giris_basarili)
```

```python
#BÖLÜM 5: math MODÜLÜ VE DAHİLİ FONKSİYONLAR

#Daire alanı ve çevresi hesaplama
import math

yaricap = 7

#Alan = π * r^2
alan = math.pi * math.pow(yaricap, 2)

#Cevre = 2 * π * r
cevre = 2 * math.pi * yaricap

print(f"Yarıçapı {yaricap} olan dairenin alanı: {alan}")
print(f"Yarıçapı {yaricap} olan dairenin çevresi: {cevre}")

#Liste Fonksiyonları (max, min, sum)
sayilar = [15, 4, 27, 13, 32, 8]

en_buyuk = max(sayilar)
en_kucuk = min(sayilar)
toplam = sum(sayilar)

print(f"Listenin en büyük elemanı: {en_buyuk}")
print(f"Listenin en küçük elemanı: {en_kucuk}")
print(f"Listedeki sayıların toplamı: {toplam}")
```
